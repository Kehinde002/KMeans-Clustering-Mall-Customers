### Importing the Dependecies

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

### Data Collection & Analysis 

```python
# loading the data from csv file to a Pandas DataFrame
customer_data = pd.read_csv("Mall_Customers.csv", header = 0, sep = ",")

print(customer_data)
```

### Choosing The Annual Income & Spending Score Column

```python
X = customer_data.iloc[:,[3,4]].values
```

```python
print(X)
```

### Choosing The Number of Clusters

##### WCSS -> Within Clusters Sum of Squares

```python
# Finding WCSS value for different numbers of clusters
# For the memory leak warning (Windows specific):
import os
os.environ["OMP_NUM_THREADS"] = "1"
import warnings
warnings.filterwarnings('ignore')

wcss = []

for i in range(1,11):
    kmeans = KMeans(n_clusters = i, init ='k-means++', random_state = 42)
    kmeans.fit(X)
    wcss.append(kmeans.inertia_)
```

```python
# Plot an elbow graph

sns.set()
plt.plot(range(1,11), wcss)
plt.title('The Elbow Point Graph')
plt.xlabel('Numbers of Clusters')
plt.ylabel('WCSS')
plt.show()
```

#### Optimal Number of Cluster = 5

### Training The K-Means Clustering Model

```python
kmeans = KMeans(n_clusters = 5, init ='k-means++', random_state = 0)

# return a a label for each data point based on their cluster
Y = kmeans.fit_predict(X)

print(Y)
```

#### 5 cluster - 0, 1 , 2, 3, 4

### Visualizing all the clusters

```python
# Plotting all the Clusters and their Centroids

plt.figure(figsize =(8,8))
plt.scatter(X[Y==0,0] ,X[Y==0,1], s=50, c='green', label='Cluster 1')
plt.scatter(X[Y==1,0] ,X[Y==1,1], s=50, c='red', label='Cluster 2')
plt.scatter(X[Y==2,0] ,X[Y==2,1], s=50, c='yellow', label='Cluster 3')
plt.scatter(X[Y==3,0] ,X[Y==3,1], s=50, c='violet', label='Cluster 4')
plt.scatter(X[Y==4,0] ,X[Y==4,1], s=50, c='blue', label='Cluster 5')

# Plot the Centroids
plt.scatter(kmeans.cluster_centers_[:,0], kmeans.cluster_centers_[:,1], s = 100 , c = 'cyan', label = 'Centroid')
plt.title('Customer Groups')
plt.xlabel('Annual Income')
plt.ylabel('Spending Score')
plt.show()
```

### Insights

######  Cluster 1 (Green): High Income, Low Spending
These customers have significant purchasing power but aren't spending it at your mall.

Insight: They might be unsatisfied with the current luxury offerings or are simply very frugal.

Recommendation: Conduct a survey to see if they want more premium brands. Offer "Concierge" services or exclusive, high-end loyalty rewards (e.g., private shopping events) to entice them to spend more.

 Cluster 2 (Red): Average Income, Average Spending
This is your largest group and represents the "average" mall-goer.

Insight: They are consistent and likely value the overall mall experience (food court, movies, etc.) rather than just shopping.

Recommendation: Use "Buy One, Get One" (BOGO) offers or standard discount coupons. Focus on mass-marketing campaigns to keep them coming back regularly.

 Cluster 3 (Yellow): High Income, High Spending
These are your most profitable customers (The "Whales").

Insight: They have the money and the willingness to spend it. They likely prefer quality and prestige.

Recommendation: Direct Marketing is key. Don't send them general discount coupons; instead, send invitations to new collection launches, offer free valet parking, or provide a dedicated personal shopper. Focus on retention at all costs.

 Cluster 4 (Violet): Low Income, High Spending
These customers spend a lot despite having lower incomes.

Insight: Likely a younger demographic or impulsive shoppers who follow trends.

Recommendation: Use social media marketing and "limited time offers." They are likely attracted to "Fast Fashion" and "Instagrammable" experiences. Focus on visual marketing and "buy now, pay later" options.

 Cluster 5 (Blue): Low Income, Low Spending
These are your least profitable customers in terms of direct sales.

Insight: They are very price-sensitive and likely only visit for essentials or window shopping.

Recommendation: Do not spend a large marketing budget on this group. Offer deep-discount clearance sales or "Value Packs" to capture what little discretionary income they have without high acquisition costs.
